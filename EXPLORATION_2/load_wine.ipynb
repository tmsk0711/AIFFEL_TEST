```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import confusion_matrix
```

```python
wine = load_wine()

wine_data = wine.data

wine_label = wine.target
wine_names = wine.feature_names
# print(wine.DESCR)

# print(wine_label)

X_train, X_test, y_train, y_test = train_test_split(wine.data,
                                                   wine_label,
                                                   test_size=0.2,
                                                   random_state=10)
print('X_train 개수: ', len(X_train),', X_test 개수: ', len(X_test))
# X_train =142, X_test=36
print("X =",y_train, y_test ) # 리스트 정상 출력

#decision_tree===============================================
decision_tree = DecisionTreeClassifier(random_state=32)

decision_tree.fit(X_train, y_train)

y_pred = decision_tree.predict(X_test)
y_pred

accuracy = accuracy_score(y_test, y_pred)
accuracy # 0.9444444444444444

# confusion_matrix(y_test, y_pred)
# classification_report(y_test, y_pred)
tree_report = classification_report(y_test, y_pred)

#random_forest===============================================
random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)

y_pred = random_forest.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
accuracy # 0.9444444444444444
forest_report = classification_report(y_test, y_pred)

#SVM Model===============================================
svm_model = svm.SVC()
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
accuracy # 0.6944444444444444
svm_report = classification_report(y_test, y_pred)

#sgd_model===============================================
sgd_model = SGDClassifier()
sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
accuracy # 0.3611111111111111
sgd_report = classification_report(y_test, y_pred)
#Logistic Regression===============================================

logistic_model = LogisticRegression()

logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
accuracy # 0.9166666666666666
logistic_report = classification_report(y_test, y_pred)


print("sgd_report :"+sgd_report+'\n'
      "logistic_report :"+logistic_report+'\n'
      "svm_report :"+svm_report+'\n'
      "forest_report :"+forest_report+'\n'
      "tree_report :"+tree_report+'\n'
     )
# forest , tree 유리 precision값이 높게나옴
```
